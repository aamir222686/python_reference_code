```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", 
                 headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
con = r.content
soup = BeautifulSoup(con, "html.parser")
```

```python
page_nr = soup.find_all('a', {'class': 'Page'})[-1].text
```

```python
base_url = "http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="
l = []
for page in range(0,int(page_nr)*10,10):
    r = requests.get(base_url+str(page)+'.html', headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    con = r.content
    soup = BeautifulSoup(con, "html.parser")
    all = soup.find_all('div', {'class': 'propertyRow'})
    for item in all:
        d = {}
        try:
            d['price'] = item.find('h4', {'class': 'propPrice'}).text.replace("\n", '').replace(" ", "").replace("$", "")
        except:
            d['price'] ='None'
        try:
            d['address'] = item.find('div', {'class': 'primaryDetails'}).text.replace("\n", '')
        except:
            d['address'] = 'None'
        try:
            d['bedroom'] = item.find('span', {'class': 'infoBed'}).text.replace("\n", '')[:1]
        except:
            d['bedroom'] = 'None'
        try:
            d['bathrooms'] = item.find('span', {'class': 'infoValueFullBath'}).text[:1]
        except:
            d['bathrooms'] = 'None'
        try:
            d['size'] = item.find('span', {'class': 'infoSqFt'}).text[:-7]
        except:
            d['size'] = 'None'
        try:
            d['half_bath'] = item.find('span', {'class': 'infoValueHalfBath'}).text[:1]
        except:
            d['half_bath'] = 'None'
        for column_group in item.find_all("div",{"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):
                    if "Lot Size" in feature_group.text:
                        d['lot_size'] = feature_name.text
        l.append(d)
```
